# Voice driven approach to use Ms word


```python
import pyttsx3
import speech_recognition as sr
import win32com.client as win32
import keyboard
import time

# Initialize the text-to-speech engine
engine = pyttsx3.init()

# Initialize the speech recognition recognizer
recognizer = sr.Recognizer()

def speak(text):
    # Function to convert text to speech
    engine.say(text)
    engine.runAndWait()

def listen():
    # Function to listen for voice input and convert it to text
    with sr.Microphone() as source:
        recognizer.adjust_for_ambient_noise(source)
        audio = recognizer.listen(source)

    try:
        text = recognizer.recognize_google(audio)
        return text.lower()
    except sr.UnknownValueError:
        speak("Sorry, I could not understand you.")
        return ""
    except sr.RequestError:
        speak("Sorry, I am currently unable to process your request.")
        return ""

def open_word():
    # Function to open Microsoft Word
    speak("Opening Microsoft Word")
    keyboard.press_and_release('ctrl+alt+w')
    word = win32.Dispatch('Word.Application')
    word.Visible = True
    doc = word.Documents.Add()
    # Wait for Word to open
    time.sleep(1)
    speak("Microsoft Word opened")

def write_document():
    # Function to start voice dictation in Word
    speak("Starting voice dictation")
    keyboard.press_and_release('win+h')
    speak("You can now start dictating in Microsoft Word.")
    speak("Press 'Windows+H' to stop")
    keyboard.wait('esc')
    speak("Dictation stopped.")


def apply_formatting():
    command = listen()
    speak("tell me what to do in formatting")
    if "bold" in command:
        bold_text()
    elif "italic" in command:
        italic_text()
    elif "underline" in command:
        underline_text()
    elif "left" in command and "justify" in command:
        left_justify_para()
    elif "right"in command and "justify"in command:
        right_justify_para()
    elif "full"in command and "justify"in command:
        full_justify_para()
    elif "center"in command and "justify"in command:
        center_justify_para()
    elif "remove"in command and "formatting" in command:
        remove_formatting()
    
    

def bold_text():
    speak("Making the text bold")
    keyboard.press_and_release('ctrl+B')
def italic_text():
    speak("Making the text italic")
    keyboard.press_and_release('ctrl+I')
def underline_text():
    speak("Making the text underline")
    keyboard.press_and_release('ctrl+U')
def left_justify_para():
    speak("Justifying paragraph left")
    keyboard.press_and_release('ctrl+L')
def right_justify_para():
    speak("Justifying paragraph right")
    keyboard.press_and_release('ctrl+R')
def full_justify_para():
    speak("Justifying full paragraph")
    keyboard.press_and_release('ctrl+J')
def center_justify_para():
    speak("Justifying paragraph in centered")
    keyboard.press_and_release('ctrl+E')
def select_next_para():
    speak("selecting next paragraph")
    keyboard.press_and_release('ctrl+shift+down')
def select_previous_para():
    speak("selecting previous paragraph")
    keyboard.press_and_release('ctrl+shift+up')
def remove_formatting():
    speak("remove formatting")
    keyboard.press_and_release('ctrl+spacebar')
def undo():
    speak("undo done")
    keyboard.press_and_release('ctrl+Z')
def redo():
    speak("redo done")
    keyboard.press_and_release('ctrl+y')
def del_previous_word():
    speak("deleting previous word")
    keyboard.press_and_release('ctrl+backspace')
def del_next_word():
    speak("deleting next word")
    keyboard.press_and_release('ctrl+delete')
def single_line_spacing():
    speak("applying single line spacing")
    keyboard.press_and_release('ctrl+1')
def double_line_spacing():
    speak("applying double line spacing")
    keyboard.press_and_release('ctrl+2')
def line_spacing_1_point_5():
    speak("applying 1 point 5 line spacing")
    keyboard.press_and_release('ctrl+5')
def add_page_number():
    speak("adding page number")  
    keyboard.press_and_release('alt+7')
    keyboard.press_and_release('enter')
    keyboard.press_and_release('enter')
    keyboard.press_and_release('alt+8')
    
def page_break():
    speak("Inserting a page break")
    keyboard.press_and_release('ctrl+enter')

def select_text_all():
    speak("Selecting all text")
    keyboard.press_and_release('ctrl+A')
    apply_formatting()

def select_right_text():
    speak("Selecting text to the right")
    keyboard.press_and_release('ctrl+shift+right')
    apply_formatting()

def select_left_text():
    speak("Selecting text to the left")
    keyboard.press_and_release('ctrl+shift+left')
    apply_formatting()  
# Main program
if __name__ == "__main__":
    speak("Hello! How can I assist you today?")

    while True:
        command = listen()

        if "open" in command and "word" in command:
            open_word()
        elif "start" in command and "writing" in command:
            write_document()
        elif "page" in command and "break" in command:
            page_break()
        elif "select" in command and "all" in command:
            select_text_all()
        elif "select" in command and "right" in command:
            select_right_text()
        elif "select" in command and "left" in command:
            select_left_text()
        elif "apply" in command and "formatting"in command:
             apply_formatting()
        elif "undo" in command:
             undo()
        elif "redo"in command:
             redo()
        elif "delete" in command and "previous" in command and "word" in command:
            del_previous_word()
        elif "delete" in command and "next" in command and "word" in command:
            del_next_word()   
        elif "single" in command and "line" in command and "spacing" in command:
            single_line_spacing()
        elif "double" in command and "line" in command and "spacing" in command:
            double_line_spacing()
        elif "line" in command and"spacing" in command and "1"in command and "point" in command and "5"in command:
            line_spacing_1_point_5()
        elif "add" in command and "page" in command and "number" in command:
            add_page_number()  
        elif "select"in command and "next"in command and "paragraph" in command:
            select_next_para()
        elif "select"in command and "previous"in command and "paragraph" in command:
            select_previous_para()
        elif "exit" in command:
            speak("Exiting the program.")
            break
```
